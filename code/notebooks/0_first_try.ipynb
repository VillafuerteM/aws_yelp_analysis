```python
# Ensure 'date' is in datetime format
df_ihop.loc[:, 'date'] = pd.to_datetime(df_ihop['date'])

# Filter for rows where state is 'FL', group by month, and count
florida_counts = df_ihop[df_ihop['state'] == 'FL'].groupby(df_ihop['date'].dt.to_period('M')).size().reset_index(name='count')

# Rename the columns for clarity
florida_counts.rename(columns={'date': 'month'}, inplace=True)
```
